1) We Do not need "Surname" and "RowNumber" to analyse data, so we drop them.
2) Notice that "Geogrpahy" and "Gender" are actually objects, we can use one hot encoding! 

```python
df_init.head()
```

```python
df_init.drop(columns=['RowNumber'], inplace=True)
df_init.head()
```

```python
surname = df_init["Surname"]
df_init.drop(columns=["Surname"], inplace=True)
df_init.head()
```

```python
#Performing one hot encoding on "Geography" and "Gender"
from sklearn import preprocessing
one_hot_encoding = preprocessing.LabelEncoder()
df_init['Geography'] = one_hot_encoding.fit_transform(df_init['Geography'])
df_init['Gender'] = one_hot_encoding.fit_transform(df_init['Gender'])
df_init.info()
```
